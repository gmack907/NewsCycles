```python
from bs4 import BeautifulSoup
from requests import get
import pandas as pd
import re

year = '2018'
month = '05'
day = '22'

url = 'https://www.realclearpolitics.com/'+year+'/'+month+'/'+day

article_sum = []

response = get(url)
html_soup = BeautifulSoup(response.text, 'html.parser')


title_art = html_soup.find_all(class_ ='post')

for i in range(len(title_art)):
    test = title_art[i]

    try:
        title = test.find_all(class_='title')
        title = title[0].text
    except IndexError:
        ''
        pass
    
    try:
        author_source = test.find_all(class_="byline")
        author = (author_source[0].text).split(',')[0]
    
    except IndexError:
        ''
        pass
    

    try:
        source_by = test.find_all(class_='source')
        source = source_by[0].text
    
    except IndexError:
        ''
        pass
    
    index = i
    date = year+month+day
    
    art_sum = [index,date,title,author,source]
    article_sum.append(art_sum)
    
df = pd.DataFrame(article_sum,columns=['index','date','title','author','source'])
print(df)
```
